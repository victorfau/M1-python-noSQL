```python
# Dans le dataset, les données sont des strings et des entiers (integer).
# Elles sont qualitatives ou quantitative discrète


import matplotlib.pyplot as plt
import pandas
import numpy


dataset = pandas.read_csv('./data/house_pricing.csv')

allLotArea = dataset['LotArea']
allSalePrice = dataset['SalePrice']

plt.scatter(allLotArea,allSalePrice)

plt.title('lot area avec saleprice')
plt.xlabel('lot area')
plt.ylabel('sale price')


```

```python
import matplotlib.pyplot as plt
import pandas
import numpy


dataset = pandas.read_csv('./data/house_pricing.csv')

allLotArea = dataset['LotArea']
allSalePrice = dataset['SalePrice']

plt.scatter(allLotArea[allLotArea<20000],allSalePrice[allSalePrice<500000])

plt.title('lot area avec saleprice')
plt.xlabel('lot area')
plt.ylabel('sale price')
```
